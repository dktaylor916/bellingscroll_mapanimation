```python
import pandas as pd
import geocoder
```

```python
df = pd.read_csv('police.csv')
```

```python
df = df[~df['Type of Violation'].isnull()]
```

```python
df = df[['City','Upload Date\n (MM/DD/YYYY)','Upload Time (CET)','Local incident date (MM/DD/YYYY)','Hyperlink','Latitude','Longitude','Type of Violation']]
```

```python
df = df[~df['Upload Date\n (MM/DD/YYYY)'].isnull()].reset_index()
```

```python
df['City']=df['City'].apply(lambda x: x+', USA')
```

```python
df
```

```python
for x in range(len(df)):
    if type(df['Latitude'][x]) is str:
        a = df['Latitude'][x].replace(',', '')
        df['Latitude'][x] = float(a)
    else:
        df['Latitude'][x] = geocoder.bing(df['City'][x],key= 'AibpFBy3GvwJRmg0kKdc-UEH6ypVoCdI9VbrfZFZEISFw11CpFgbJhou9xgBZooL').latlng[0]
        df['Longitude'][x] = geocoder.bing(df['City'][x],key= 'AibpFBy3GvwJRmg0kKdc-UEH6ypVoCdI9VbrfZFZEISFw11CpFgbJhou9xgBZooL').latlng[1]
        
```

```python
df['date'] = pd.to_datetime(df['Upload Date\n (MM/DD/YYYY)'],format = ('%m/%d/%Y'))
```

```python
df = df.sort_values(['date','Upload Time (CET)'])
```

```python
del df['index']
```

```python
df=df.reset_index()
```

```python
df['index'] = list(df.index)
```

```python
df
```

```python
time = pd.DataFrame(list(df['date'].unique()),columns=['date']).reset_index().rename(columns={'index':'date_index'})
```

```python
df = df.merge(time, how='left', on='date')
```

```python
df.to_csv(r'police_clean3.csv')
```
